```python
# Dataset shows the refugees and asylum seekers coming to Bulgaria between 1992 and 2017

import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# load refugees to Bulgaria csv

refugees = pd.read_csv('unhcr-time-series-residing-bgr-csv.csv')
refugees.head(5)
```

```python
refugees.info()
```

```python
# remove NaN values from refugees and first row of data
refugees.dropna(subset=['Value'], inplace=True)
refugees.drop(refugees.index[0], inplace=True)
refugees.head()
```

```python
# rename columns of the dataset
refugees.columns = ['year', 'country_territory_of_asylum', 'origin', 'population_type', 'value']
refugees.head()
```

```python
# print the length of the 4 population types
refugees['population_type'].value_counts()
```

```python
# change columns value and year to int
refugees['year'] = refugees['year'].astype('int32')
refugees['value'] = refugees['value'].astype('int32')
refugees.head()
```

```python
refugees.dtypes
```

```python
refugees.describe()
```

```python
# calculating a total per year from each country
refugees_types_merged = refugees.groupby(['year','origin']).sum()['value'].reset_index()
refugees_types_merged.head(5)
```

```python
# print the names of the countries
unique_countries = refugees.origin.unique().tolist()
len(unique_countries)
unique_countries.sort()
print(unique_countries)
```

```python
# Rename Congo (as per official UN list of countries)
refugees = refugees.replace('Dem. Rep. of the Congo', 'Republic of the Congo')
refugees_types_merged = refugees_types_merged.replace('Dem. Rep. of the Congo', 'Republic of the Congo')
```

```python
# Rename Guinea (as per official UN list of countries)
refugees = refugees.replace('Guinea', 'Papua New Guinea')
refugees_types_merged = refugees_types_merged.replace('Guinea', 'Papua New Guinea')
```

```python
# rename Macao to include * for  explanation
refugees = refugees.replace('China, Macao SAR', 'China, Macao SAR*')
refugees_types_merged = refugees_types_merged.replace('China, Macao SAR', 'China, Macao SAR*')
```

```python
# rename Israel to include * for explanation
refugees = refugees.replace('Israel', 'Israel*')
refugees_types_merged = refugees_types_merged.replace('Israel', 'Israel*')
```

```python
# rename Serbia to include ** for explanation
refugees = refugees.replace('Serbia and Kosovo (S/RES/1244 (1999))', 
                            'Serbia and Kosovo (S/RES/1244 (1999))*')
```

```python
# rename Western Sahara to include * for explanation
refugees = refugees.replace('Western Sahara', 'Western Sahara*')
refugees_types_merged = refugees_types_merged.replace('Western Sahara', 'Western Sahara*')
```

```python
# rename Palestinian to include * for explanation
refugees = refugees.replace('Palestinian', 'State of Palestine*')
refugees_types_merged = refugees_types_merged.replace('Palestinian', 'State of Palestine*')
```

```python
# Special cases
# *SAR - special administrative region
# *Israel - (UN website)
# Palestinian
# *Serbia and Kosovo (S/RES/1244 (1999))
# *Western Sahara - Non-self governing territory
```

```python
# rename Turkey to include * for explanation
refugees = refugees.replace('Turkey', 'Turkey*')
refugees_types_merged = refugees_types_merged.replace('Turkey', 'Turkey*')
```

```python
# create lists for the different regions as per United Nations Regional Groups of Member 
# States and create function to assign values to each country

african = ['Algeria', 'Angola', 'Benin', 'Burkina Faso', 'Burundi', 'Cabo Verde', 'Cameroon',
          'Central African Rep.', 'Chad', 'Comoros', 'Congo', "Cote d'Ivoire", 'Republic of the Congo',
          'Egypt', 'Eritrea', 'Ethiopia', 'Gabon', 'Gambia', 'Ghana', 'Kenya', 'Liberia', 'Libya',
          'Mali', 'Mauritania', 'Morocco', 'Mozambique', 'Niger', 'Nigeria', 'Rwanda', 'Senegal',
          'Sierra Leone', 'Somalia', 'South Africa', 'Sudan', 'Togo', 'Tunisia', 'Uganda',
          'United Rep. of Tanzania', 'Western Sahara*', 'Zambia', 'Zimbabwe']
asia_pacific = ['Afghanistan', 'Bahrain', 'Bangladesh', 'China', 'China, Macao SAR*', 'Papua New Guinea',
               'India', 'Iran (Islamic Rep. of)', 'Iraq', 'Jordan', 'Kazakhstan', 'Kuwait', 
               'Kyrgyzstan', "Lao People's Dem. Rep.", 'Lebanon', 'Malaysia', 'Mongolia', 
               'Myanmar', 'Nepal', 'Pakistan', 'State of Palestine*', 'Philippines', 'Singapore',
               'Sri Lanka', 'Syrian Arab Rep.', 'Tajikistan', 'Turkey*', 'Turkmenistan', 
               'United Arab Emirates', 'Uzbekistan', 'Viet Nam', 'Yemen']
eastern_european = ['Albania', 'Armenia', 'Azerbeijan', 'Belarus', 'Bosnia and Herzegovina',
                   'Georgia', 'Lithuania', 'Poland', 'Rep. of Moldova', 'Russian Federation',
                   'Serbia and Kosovo (S/RES/1244 (1999))*', 'The former Yugoslav Republic of Macedonia',
                   'Ukraine', ]
latin_american = ['Chile', 'Colombia', 'Cuba', 'Dominican Rep.', 'Ecuador', 'Peru', 'Uruguay',
                 ]
western_european = ['Australia', 'Germany', 'Israel*']
other = ['Stateless', 'Various/Unknown']


def add_region(row):
    if row['origin'] in african:
        val = 'African Group'
    elif row['origin'] in asia_pacific:
        val = 'Asia-Pacific Group'
    elif row['origin'] in eastern_european:
        val = 'Eastern European Group'
    elif row['origin'] in latin_american:
        val = 'Latin American and Caribbean Group (GRULAC)'
    elif row['origin'] in western_european:
        val = 'Western European and Others Group (WEOG)'
    else:
        val = 'Other / Not specified'
    return val
```

```python
# add the new column to the refugees dataset
refugees['region_groups'] = refugees.apply(add_region, axis=1)
```

```python
# sort year in ascending order
refugees = refugees.sort_values(['year'], ascending=True)
refugees.head(5)
```

```python
# calculating total amount of refugees per year
value_per_year = refugees.groupby(['year']).sum()[['value']].reset_index()
#print(value_per_year)
plt.rcParams['figure.figsize'] = [15, 6]
plt.locator_params(axis='x', nbins=20)

plt.bar( value_per_year['year'], value_per_year['value'], align='center', color='#999999' )


plt.xlabel('Year')
plt.ylabel('Refugees')
plt.title('Refugees per year')

# plt.savefig('ref_per_year.png')
```

```python
# substract the amount of refugees, asylum seekers, returned refugees per year
types_per_year = refugees.groupby(['year', 'population_type']).sum()[['value']].reset_index()
types_per_year.population_type.unique()

plt.rcParams['figure.figsize'] = [10, 10]
colors = ['olivedrab', 'khaki', 'lightblue', 'palegreen']
refugees.groupby(['year', 'population_type']).sum()[['value']].unstack('population_type').plot.bar(stacked=True, color=colors)

plt.title('Refugees by year and type')

# plt.savefig('ref_by_year_and_type_stacked.png')
```

```python
# grouped bar chart for types_per_year
colors = ['olivedrab', 'khaki', 'lightblue', 'palegreen']
refugees.groupby(['year', 'population_type']).sum()[['value']].unstack('population_type').plot.bar(color=colors)

plt.title('Refugees by year and type')

# plt.savefig('ref_by_year_and_type_grouped.png')
```

```python
# scatter plot showing the amount of refugees and year
plt.figure(figsize=(15, 15))


plt.scatter( refugees['year'], refugees['value'], alpha=0.5)
```

```python
# plot line chart by the origin of the refugees
fig, ax = plt.subplots(figsize=(40,40))
refugees.groupby(['year','origin']).sum()['value'].unstack().plot(ax=ax, linewidth=3)
# place legend below the chart
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), fancybox=True, 
           shadow=True, ncol=5, borderaxespad=0., prop={'size': 25})

plt.title('Refugees coming to Bulgaria')

# plt.savefig('ref_line.png')
```

```python
# calculating total number of refugees by country, sorting in descending order
origin_and_value = refugees.groupby('origin').sum()[['value']]
origin_and_value = origin_and_value.sort_values(['value'], ascending=False)
origin_and_value.head(5)
```

```python
# substract the 3 outliers from the main dataset
afg = refugees[refugees['origin'] == 'Afghanistan']
iq = refugees[refugees['origin'] == 'Iraq']
sar = refugees[refugees['origin'] == 'Syrian Arab Rep.']
```

```python
# join the 3 outliers in a new dataset
outliers = pd.concat([afg,iq,sar])
outliers.head()
```

```python
outliers.describe()
```

```python
# grouped bar chart showing the outlier countries' refugees by type
outliers.groupby(['year', 'origin']).sum()[['value']].unstack('origin').plot.bar()

plt.title('Refugees from Afganistan, Iraq and Syria')

# plt.savefig('outliers_grouped_by_type.png')
```

```python
# plot line chart by the origin of the refugees
fig, ax = plt.subplots(figsize=(20,10))
outliers.groupby(['year', 'origin']).sum()[['value']].unstack().plot(ax=ax, linewidth=5)
# place legend on the right side of the line chart
plt.legend(loc=2, prop={'size': 10})

plt.title('Refugees from Afganistan, Iraq and Syria')

# plt.savefig('outliers_line.png')
```

```python
# stacked bar plot of the outliers based on the type of refugees
plt.rcParams['figure.figsize'] = [10, 10]
outliers.groupby(['year', 'population_type']).sum()[['value']].unstack('population_type').plot.bar(stacked=True)

plt.title('Total amount of refugees from the outlier countries by population type')

# plt.savefig('outliers_grouped_by_type.png')
```
