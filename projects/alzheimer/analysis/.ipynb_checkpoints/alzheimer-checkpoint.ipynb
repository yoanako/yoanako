```python
# Dataset represents over 2000 clinical trials for Alzheimer's disease
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
```

```python
# set colors and labels for interventional and observational studies
color_int = '#ff6961'
color_obs = '#8bd6f2'
colors = [color_int, color_obs]
labels = ['Interventional', 'Observational']
```

```python
# load alzheimer dataset
alzheimer = pd.read_csv('alzheimer.csv')
alzheimer.head(1)
```

```python
# show all columns
alzheimer.columns
```

```python
# clear not needed data
del alzheimer['Rank']
del alzheimer['Acronym']
del alzheimer['Study Results']
del alzheimer['Interventions']
del alzheimer['Outcome Measures']
del alzheimer['Sponsor/Collaborators']
del alzheimer['Funded Bys']
del alzheimer['Study Designs']
del alzheimer['Other IDs']
del alzheimer['Primary Completion Date']
del alzheimer['First Posted']
del alzheimer['Results First Posted']
del alzheimer['Last Update Posted']
del alzheimer['Locations']
del alzheimer['Study Documents']

alzheimer.head(1)
```

```python
# show unique values in Age column
alzheimer.Age.unique()
```

```python
# create column clean_age which contains a clean string of the min and max age of study subjects

alzheimer['clean_age'] = alzheimer.Age.str.split('\xa0').str[0]
alzheimer['clean_age'].value_counts()
```

```python
alzheimer.head(1)
```

```python
# create Age groups to subset dataset
# 1. up to 18
# 2. 18 to 40
# 3. 40 to 65
# 4. 65 to 85
# 5. 85 and older
```

```python
# parse the Age column to extract the min age of each study
alzheimer['min_age'] = alzheimer['clean_age'].str[:2]
alzheimer.head(1)
```

```python
# count unique values and their quantities in min_age column
alzheimer['min_age'].value_counts()
```

```python
# parse the min_age column to conver 'ch' and 'up' to NaN
alzheimer['min_age'] = alzheimer.min_age.str.extract('(\d+)')
# print(age_starting)
```

```python
# replace 'Child, Adult, Older Adult' with 0 in min_age column (as they include all ages)
alzheimer['min_age'].fillna(0, inplace=True)
```

```python
# convert min_age column to int32
alzheimer['min_age'] = alzheimer['min_age'].astype('int32')
alzheimer.dtypes
```

```python
# create new column and group ages in it
conditions = [
    (alzheimer['min_age'] <= 18) & (alzheimer['min_age'] >= 1 ),
    (alzheimer['min_age'] > 18) & (alzheimer['min_age'] <= 40),
    (alzheimer['min_age'] > 40) & (alzheimer['min_age'] <= 65),
    (alzheimer['min_age'] > 65),
    (alzheimer['min_age'] == 0)
]
     
choices = ['up to 18 years old',
          '18 to 40 years old',
          '40 to 65 years old',
          '65 and older',
          'Age not specified']
alzheimer['age_groups'] = np.select(conditions, choices)
alzheimer.head()
```

```python
# count unique values and their quantities in age_groups column
alzheimer['age_groups'].value_counts()
```

```python
# count unique values and their quantities in Status column
alzheimer['Status'].value_counts()
```

```python
alzheimer.info()
```

```python
# check for NaN values
null_columns = alzheimer.columns[alzheimer.isnull().any()]
alzheimer[null_columns].isnull().sum()
```

```python
# remove rows with NaN in columns 'Enrollment', 'Gender'
alzheimer = alzheimer.dropna(subset=['Enrollment', 'Gender'])
alzheimer.head(5)
```

```python
# convert Enrollment type to int32
alzheimer['Enrollment'] = alzheimer['Enrollment'].astype('int32')
alzheimer.dtypes
```

```python
# replace all NaN values from column Phases (observational studies) with 'Phase 4'
alzheimer['Phases'].fillna('Phase 4', inplace=True)
alzheimer.head()
```

```python
# create a clean dataset with the columns which are going to be used
alzheimer_clean = alzheimer.loc[:, ['NCT Number', 'Title', 'Status', 'clean_age', 'Phases', 'Enrollment', 
                                    'Study Type', 'URL', 'age_groups']]
```

```python
# get clean dataset with the columns needed for visualizations
alzheimer_clean.columns = ['nct_number', 'title', 'status', 'clean_age', 'phases', 'enrollment', 'study_type', 
                                   'url', 'age_groups']
alzheimer_clean.head()
```

```python
# conver the dataframe to json
json_data = alzheimer_clean.to_dict(orient='records')
print(json_data)
```

```python
# nest the json
    
formatted = {'name': 'Alzheimer', 'children': []}
age_groups = []
status = []

for d in json_data:
    d['name'] = d['nct_number']

    # locate and append unique age_groups values to the age_groups list
    if d['age_groups'] not in age_groups:
        age_groups.append(d['age_groups'])

        # append structure for the nested json (formatted dict)
        formatted['children'].append({'name': d['age_groups'], 'children': []})

for d in json_data:
    if d['status'] not in status:
        # locate and append unique statuses to the status list
        status.append(d['status'])

        for age in formatted['children']:
            # append the status values to the 'children' key in the formatted dict
            age['children'].append({'name': d['status'], 'children': []})

    for age in formatted['children']:
        # for Observational studies only (group by study type)
        if d['study_type'] == 'Observational' and age['name'] == d['age_groups']:
            for s in status:
                # check that current data point has a specific status
                if s == d['status']:
                    for f in age['children']:
                        # add data to unique status group
                        if f['name'] == s:
                            f['children'].append(d)                   
                            
        # for Interventional studies only (group by study type)                    
        if d['study_type'] == 'Interventional' and age['name'] == d['age_groups']:
            for s in status:
                # check that current data point has a specific status
                if s == d['status']:
                    for f in age['children']:
                        # add data to unique status group
                        if f['name'] == s:
                            f['children'].append(d)

print(formatted)
```
